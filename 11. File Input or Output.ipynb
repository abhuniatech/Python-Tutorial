# 11. File Input or Output

## Types of Files
There are 2 types of files:
1. Text files (.txt, .c etc)
2. Binary files (.jpg, .dat etc)

Python has a lot of functions for reading, updating and deleting files.

## Opening a file
Python has an open() function for opening files. It takes 2 parameters: filename and mode

```python
# Use open function to read the content of a file
f=open('sample.txt', 'r')
data = f.read() # reads its contents
print(data) # prints its contents
f.close() #closes the file 
```

```python
f.read(2) # reads first 2 charaters
```

```python
f=open('sample.txt', 'r')
data = f.readline() # reads first line
print(data)
data = f.readline() # reads second line
print(data)
data = f.readline() # reads third line and so on
print(data)                                                                      
f.close() #closes the file 
```

## Writing files in python
In order to write a file, we first open it in 'write' or 'append' mode after which, we use the python's f.write() method to write to the file!
