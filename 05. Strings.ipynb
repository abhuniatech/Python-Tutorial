## 5.3 String indexing

```python
name = "Arindam"
```

```python
print(name[5]) #prints 5th element in a string
```

```python
name[6]="d" #element of the string cannot be changed 
```

## 5.4 String Slicing
A string in a python can be sliced for getting a part of the string


```python
print(name[0:5]) # prints elements from 0th to (5-1)th element of the string
```

```python
print(name[2:]) # prints elements from 2nd to last element
```

```python
print(name[:5]) #prints elements from initial to (5-1)th element
```

## 5.5 Negative indexing in python 

```python
print(name[-4:-1]) # prints elelemnts from (-4)th to ((-1)-1)th element
```

```python
print(name[-7:-1:2]) #prints elements from 1st to (4-1)th with 2 gaps in between
```
